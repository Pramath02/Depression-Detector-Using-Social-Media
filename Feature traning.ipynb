```python
data2 =data.iloc[:,2:]
# data=data.drop(['ItemID'], axis = 1) 
data2=data2.drop(['SentimentSource'], axis = 1) 
data2=data2.drop(['SentimentText'], axis = 1) 


```

```python
X=data2.iloc[:,1:].values
y=data2.iloc[:,0].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split;
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=9,random_state=45,min_samples_split=5)
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
feat_labels=['Analytic','Clout','Authentic','Tone','pronoun','ppron','i','we','you','shehe','they','ipron','article','prep','auxverb','adverb','conj','negate','verb','adj','compare','interrog','posemo','negemo','anx','anger','sad','social','family','friend','focuspast','focuspresent','focusfuture']
for feature in zip(feat_labels, clf.feature_importances_):
    print(feature)
```

```python
sfm = SelectFromModel(clf, threshold=0.01)
```

```python
sfm.fit(X_train, y_train)
```

```python
for feature_list_index in sfm.get_support(indices=True):
    print(feat_labels[feature_list_index])
```

```python
X_important_train = sfm.transform(X_train)
X_important_test = sfm.transform(X_test)
```

```python
clf_important =RandomForestClassifier(n_estimators=9,random_state=45,min_samples_split=5)
```

```python
clf_important.fit(X_important_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# Apply The Full Featured Classifier To The Test Data
y_important_pred = clf_important.predict(X_important_test)

# View The Accuracy Of Our Limited Feature (2 Features) Model
accuracy_score(y_test, y_important_pred)
```
